```python
import numpy as np
import matplotlib.pyplot as plt

# Load the dataset
dataset = np.load('ising_dataset_2.npz')
data = dataset['data']
labels = dataset['labels']

# Set up the grid parameters
n_rows = 10  # Number of different beta values to show
n_cols = 10  # Number of examples per beta value
fig, axes = plt.subplots(n_rows, n_cols, figsize=(15, 15))

# Get unique beta values
unique_betas = np.unique(labels)
selected_betas = unique_betas[:n_rows]  # Select first n_rows beta values

# Plot the examples
for i, beta in enumerate(selected_betas):
    # Get indices for current beta value
    beta_indices = np.where(labels == beta)[0]
    # Randomly select n_cols samples
    selected_indices = np.random.choice(beta_indices, n_cols, replace=False)
    
    for j in range(n_cols):
        ax = axes[i, j]
        im = ax.imshow(data[selected_indices[j]], cmap='binary')
        ax.axis('off')
        
        # Add beta value as title only for the first column
        if j == 0:
            ax.set_title(f' = {beta:.2f}', pad=10)

plt.tight_layout()
plt.show()
```
